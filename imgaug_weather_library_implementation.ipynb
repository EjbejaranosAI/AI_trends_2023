## imgaug.augmenters.weather


https://imgaug.readthedocs.io/en/latest/source/api_augmenters_weather.html#imgaug.augmenters.weather.FastSnowyLandscape

````python
import cv2
import imgaug.augmenters as iaa
import matplotlib.pyplot as plt
from imgaug.augmenters.weather import CloudLayer

# Load the original image
path_img = "original.png"
img = cv2.imread(path_img)
original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Create the augmenters
cloud_layer = iaa.CloudLayer(255,[-3.5, -1],4,0.3,0.4,10,-2.5,1.1,1.2)
rain_augmenter = iaa.Rain()
clouds_original = iaa.Clouds()
SnowflakesLayer = iaa.SnowflakesLayer((0.01, 0.075), (0.5,0.3), (0.2, 0.9), 0.5, (-2, 20), (0.001, 0.03), (0.0001, 0.001), blur_sigma_limits=(0.5, 3.9), seed=None, name=None, random_state='deprecated', deterministic='deprecated')

# Apply the augmenters to the original image
cloud = cloud_layer(image=original)
rain = rain_augmenter(image=original)
cloud_img = clouds_original(image=original)
snow_layer = SnowflakesLayer(image=original)

# Create a grid to display the images
fig, axs = plt.subplots(2, 2)
fig.suptitle("Weather Transformations")

# Display the images in the grid
axs[0, 0].imshow(original)
axs[0, 0].set_title("Original")
axs[0, 1].imshow(cloud)
axs[0, 1].set_title("Cloud Layer")
axs[1, 0].imshow(rain)
axs[1, 0].set_title("Rain")
axs[1, 1].imshow(snow_layer)
axs[1, 1].set_title("Snowflakes Layer")

# Remove the axis labels
for ax in axs.flat:
    ax.axis("off")

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()




````
